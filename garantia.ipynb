```python
import pandas as pd
import warnings
import os

warnings.filterwarnings("ignore")
```

```python
df_garantia = pd.read_excel("docs/garantia.xlsx")
df_garantia = df_garantia.sort_values(by=['DEUDOR', 'PERIODO'])
df_garantia['PERIODO'] = df_garantia['PERIODO'].astype("Int64")
df_garantia['DEUDOR'] = df_garantia['DEUDOR'].astype("Int64")
df_garantia['ACREEDOR'] = df_garantia['ACREEDOR'].astype("Int64")
df_garantia.reset_index(drop=True, inplace=True)
df_garantia.head(5)
```

```python
deudores_con_garantia = df_garantia.groupby('DEUDOR')['GARANTIA_FINAL'].transform('sum') != 0
df_garantia = df_garantia[deudores_con_garantia]
```

```python
df_garantia['DIFERENCIA'] = df_garantia.groupby('DEUDOR')['GARANTIA_FINAL'].diff()
```

```python
deudores_con_diferencia = df_garantia.groupby('DEUDOR')['DIFERENCIA'].transform('sum') != 0
df_garantia = df_garantia[deudores_con_diferencia]
```

```python
df_garantia.to_excel("docs/garantia_diferencias.xlsx", index=False)
```

```python
os.startfile("docs/garantia_diferencias.xlsx")
```
